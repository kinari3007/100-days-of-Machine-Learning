### 0D Tensor : Scalar
Real world - A single temperature reading from a weather sensor

```python
import torch

temperature = torch.tensor(36.6)

print(temperature)        # tensensor(36.6000)
print(temperature.ndim)   # 0 -> no dimensions, just a single value
print(temperature.shape)  # torch.Size([]) -> empty shape, since it's a scalar value
print(temperature.item())  # 36.6 -> get the Python number from the tensor
```

What's happening:
36.6  ← just one number, no rows, no columns, nothing

> Single sensor reading, single loss value after training, single accuracy score — all scalars.

### 1D Tensor : Vector
Real world - One patient's medical report with multiple readings

```python
patient_record = torch.tensor([98.6, 120.0, 80.0, 95.0])
#                             temp, bp_sys, bp_di, oxygen

print(patient_record)        # tensor([ 98.6000, 120.0000,  80.0000,  95.0000])
print(patient_record.ndim)   # 1 -> one dimension, a list of values
print(patient_record.shape)  # torch.Size([4]) -> 4 values in the tensor
print(patient_record[0])     # tensor(98.6000) -> first value (temperature)
print(patient_record[1])     # tensor(120.0000) -> second value (blood pressure systolic)
print(patient_record[2])     # tensor(80.0000) -> third value (blood pressure diastolic)
print(patient_record[3])     # tensor(95.0000) -> fourth value (oxygen saturation)  
```

- What's happening:
→ 4 features of ONE patient = 1D tensor of size 4

> One row of a dataset, one word embedding, one feature vector — all 1D tensors.

### 2D Tensor : Matrix
Real world - A small hospital dataset - multiple patients, multiple readings

```python
hospital_data = torch.tensor([
    [98.6, 120.0, 80.0, 95.0],   # patient 1
    [99.1, 130.0, 85.0, 93.0],   # patient 2
    [97.8, 115.0, 75.0, 97.0],   # patient 3
])

print(hospital_data.ndim)   # 2  → rows and columns
print(hospital_data.shape)  # torch.Size([3, 4]) → 3 patients, 4 readings each
print(hospital_data[0])     # tensor([ 98.6000, 120.0000,  80.0000,  95.0000]) → first patient's data
print(hospital_data[1])     # tensor([ 99.1000, 130.0000,  85.0000,  93.0000]) → second patient's data
print(hospital_data[2])     # tensor([ 97.8000, 115.0000,  75.0000,  97.0000]) → third patient's data
```

What's happening : 
→ 3 rows × 4 columns = 2D tensor of shape [3, 4]

> Any tabular dataset, any spreadsheet, weight matrix in neural network — all 2D tensors.

### 3D Tensor : Cube
Real world - A week of hourly stock prices for multiple companies

```python
stock_data = torch.tensor([
    # Day 1: [Apple, Google, Tesla] prices at each hour
    [[180.0, 140.0, 250.0],   # hour 1
     [182.0, 141.0, 253.0],   # hour 2
     [179.0, 139.0, 248.0]],  # hour 3

    # Day 2
    [[183.0, 142.0, 255.0],
     [185.0, 143.0, 258.0],
     [181.0, 140.0, 252.0]],

    # Day 3
    [[178.0, 138.0, 247.0],
     [180.0, 140.0, 250.0],
     [182.0, 141.0, 253.0]],
])

print(stock_data.ndim)   # 3
print(stock_data.shape)  # torch.Size([3, 3, 3])
#                                       ↑  ↑  ↑
#                                     days hrs companies
print(stock_data[0])     # Day 1 data
print(stock_data[1])     # Day 2 data
print(stock_data[2])     # Day 3 data
```

> Time series data, batch of sentences (NLP), sequence data — all 3D tensors.

### 4D Tensor : A stack of 3D Tensors
Real world : A batch of color images fed into a neural network

```python
# Simulating a batch of 4 color photos (like Instagram photos)
# Each photo: 3 color channels (RGB), 64×64 pixels

batch_photos = torch.rand(4, 3, 64, 64)
#                          ↑  ↑   ↑  ↑
#                        imgs ch  H  W

print(batch_photos.ndim)   # 4
print(batch_photos.shape)  # torch.Size([4, 3, 64, 64])

# Accessing one photo
single_photo = batch_photos[0]
print(single_photo.shape)  # torch.Size([3, 64, 64]) → one image

# Accessing one channel of that photo
red_channel = batch_photos[0][0]
print(red_channel.shape)   # torch.Size([64, 64]) → just red values
```

## Accessing data inside 4D Tensor

```python
batch = torch.rand(4, 3, 64, 64)

# Get entire image 1
print(batch[0])
# shape → [3, 64, 64]  ← one full color image

# Get RED channel of image 1
print(batch[0][0])
# shape → [64, 64]  ← red pixel grid of image 1

# Get GREEN channel of image 1
print(batch[0][1])
# shape → [64, 64]  ← green pixel grid of image 1

# Get BLUE channel of image 1
print(batch[0][2])
# shape → [64, 64]  ← blue pixel grid of image 1

# Get pixel at row 10, col 20 of RED channel of image 1
print(batch[0][0][10][20])
# shape → scalar  ← single pixel value
```

### 5D Tensor : Batch  of Videos
Real world - A batch of video clips - like reels being processed by a video model

```python
# 2 video clips
# Each clip: 10 frames, 3 RGB channels, 64×64 pixels

batch_videos = torch.rand(2, 10, 3, 64, 64)
#                          ↑   ↑  ↑   ↑  ↑
#                        vids frm ch   H  W

print(batch_videos.ndim)   # 5
print(batch_videos.shape)  # torch.Size([2, 10, 3, 64, 64])

# Accessing one video
video1 = batch_videos[0]
print(video1.shape)        # torch.Size([10, 3, 64, 64]) → 10 frames

# Accessing one frame of that video
frame1 = batch_videos[0][0]
print(frame1.shape)        # torch.Size([3, 64, 64]) → one frame = one image

# Accessing red channel of that frame
red = batch_videos[0][0][0]
print(red.shape)           # torch.Size([64, 64]) → pixel grid
```
