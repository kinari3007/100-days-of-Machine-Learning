## Instance Based Learning - KNN
(memorizes, then finds the nearest neighbor)

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Training Data
X_train = np.array([[1,40],[2,50],[6,85],[7,90],[8,95],[9,100]])
y_train = np.array([0,0,1,1,1,1])  # 0: Not Fit, 1: Fit

# KNN just stores this data - it doesn't learn a model
model = KNeighborsClassifier(n_neighbors=2)
model.fit(X_train, y_train)

# Prediction : it findds the 2 nearest neighbors and takes a majority vote
print(model.predict([[5,80]]))  # Should predict 1 (Fit)

print(model.predict([[3,45]]))  # Should predict 0 (Not Fit)
```

- Here since 5 is closer to 6 and 7 therefore it given 1 as an answer
- And 0 is closer to 1 and 2 therefore it gives 0 as an answer

## Model Based Learning - Linear Regression 
(Learns formula and hidden pattern inside the data, and then forgets the data)

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Training data
X_train = np.array([[1],[2],[3],[4],[5]])  # 2D array for sklearn
y_train = np.array([2,4,6,8,10])  # Linear relationship: y = 2x 

# Linear Regression LEARNS a formula from given data
model = LinearRegression()
model.fit(X_train, y_train)

#  The formula it learned 
print(f"Formula learned: y = {model.coef_[0]}x + {model.intercept_}")  # Should be close to "2x + 0"

# Prediction using the learned formula - Here original data is not needed anymore
print(model.predict([[6]]))  # Should predict close to 12 (2*6)
```

## Both model side by side


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
import numpy as np

# Features: [action_rating, romance_rating]
X = np.array([[8,2],[7,3],[2,8],[3,9],[5,5],[6,4]])
y = np.array([1,1,0,0,1,1])  # 1=likes action, 0=likes romance

# Instance Based — KNN
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X, y)
print("KNN prediction:", knn.predict([[7,2]]))      # finds similar users

# Model Based — Logistic Regression
lr = LogisticRegression()
lr.fit(X, y)
print("LR prediction:", lr.predict([[7,2]]))        # applies learned formula
```
